```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os
from pathlib import Path

# Set style and display options
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (14, 7)
pd.set_option('display.max_columns', None)
%matplotlib inline

# Load the data
data_path = Path('data/processed/final_crop_soil_data_with_yield_fixed.csv')
df = pd.read_csv(data_path)

# Display basic information
print("Data Shape:", df.shape)
df.head()
```

```python
# Data Exploration
print("\\n=== Basic Data Information ===")
print("\\nData Types and Missing Values:")
print(df.info())

print("\\n\\n=== Summary Statistics ===")
print(df.describe())

# Check for missing values
print("\\n\\n=== Missing Values ===")
print(df.isnull().sum().sort_values(ascending=False))
```

```python
# Top 10 crops by average yield
plt.figure(figsize=(14, 8))
top_crops = df.groupby('crop')['yield_value'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_crops.values, y=top_crops.index)
plt.title('Top 10 Crops by Average Yield (Tonnes/Hectare)')
plt.xlabel('Average Yield (Tonnes/Hectare)')
plt.ylabel('Crop')
plt.tight_layout()
plt.show()
```

```python
# Seasonal yield patterns
plt.figure(figsize=(14, 6))
seasonal_yield = df.groupby('season')['yield_value'].mean().sort_values()
sns.barplot(x=seasonal_yield.index, y=seasonal_yield.values)
plt.title('Average Crop Yield by Season')
plt.ylabel('Average Yield (Tonnes/Hectare)')
plt.xlabel('Season')
plt.show()
```

```python
# Weather impact on yield
weather_cols = ['avg_temp', 'total_precip', 'avg_humidity', 'avg_windspeed']
plt.figure(figsize=(16, 10))
for i, col in enumerate(weather_cols, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(data=df, x=col, y='yield_value', alpha=0.5)
    plt.title(f'Yield vs {col.replace("_", " ").title()}')
plt.tight_layout()
plt.show()
```

```python
# Soil type analysis
soil_cols = [col for col in df.columns if 'level' in col.lower()]

plt.figure(figsize=(16, 8))
for i, col in enumerate(soil_cols, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(data=df, x=col, y='yield_value')
    plt.title(col.replace('_level', '').title())
    plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Interactive yield map by state (if you have geographical data)
if 'state_name' in df.columns:
    state_yield = df.groupby('state_name')['yield_value'].mean().reset_index()
    fig = px.choropleth(
        state_yield,
        locations='state_name',
        locationmode='country names',
        color='yield_value',
        hover_name='state_name',
        color_continuous_scale='Viridis',
        title='Average Crop Yield by State'
    )
    fig.show()
```
